```python
import os, sys
import csv
import time
import pandas as pd
import numpy as np
import torch

import fastai
from fastai import vision

import util
```

#### Data
This is probably the hard part

```python
# Setup
path = vision.Path('./exp')
path.mkdir(parents=True, exist_ok=True)
image_path = vision.Path('/home/nlp/other/ML_GCN/data/voc/VOCdevkit/VOC2007/JPEGImages')
```

```python
# GPU
device = torch.device('cuda:0') # cuda:0
fastai.torch_core.defaults.device = device
```

```python
# Load preprocessed labels
csv_file = '/datadrive/datasets/multi-label-classification/voc2007/preprocessed/files/VOC2007/classification_trainval.csv'
object_categories = ['aeroplane', 'bicycle', 'bird', 'boat',
                     'bottle', 'bus', 'car', 'cat', 'chair',
                     'cow', 'diningtable', 'dog', 'horse',
                     'motorbike', 'person', 'pottedplant',
                     'sheep', 'sofa', 'train', 'tvmonitor']
df_dtypes = {category: int for category in object_categories}
df_dtypes['name'] = str
df = pd.read_csv(csv_file, dtype=df_dtypes)
df.head(2)
```

```python
# Convert to fastai format
df['classes'] = df.apply(lambda row: " ".join(c for c in object_categories if row[c]==1), axis=1)
df = df[['name', 'classes']]
df.head(3)
```

```python
# Note that the average image has only 1.5 classes
sum(len(x.split(' ')) for x in list(df['classes'])) / len(df['classes'])
```

```python
# Data augmentation 
tfms = vision.get_transforms()
```

```python
# Train test split
## In the original paper they val with the test set - that's nonsense.
np.random.seed(42)
src = vision.ImageList.from_df(df=df, cols='name', suffix='.jpg', path=image_path) \
        .split_by_rand_pct(0.2) \
        .label_from_df(label_delim=' ')
```

```python
# Dataloader
data = src.transform(tfms, size=128) \
        .databunch(bs=64) \
        .normalize(vision.imagenet_stats)
```

```python
# Example
data.show_batch(rows=2, figsize=(12,9))
```

#### Model

```python
# Model
arch = vision.models.resnet34
```

```python
# Metrics
from functools import partial
acc_with_thresh = partial(vision.accuracy_thresh, thresh=0.5)
f1_with_thresh = partial(vision.fbeta, thresh=0.2, beta=1)
learn = vision.cnn_learner(data, arch, metrics=[acc_with_thresh, f1_with_thresh, util.AveragePrecisionCallback()])
```

#### Training

```python
learn.summary()
```

```python
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, slice(1e-2))
```

```python
# learn.save('stage-1-64')
learn.unfreeze()
```

```python
learn.lr_find()
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, slice(2e-5))
```

```python
learn.freeze()
```
